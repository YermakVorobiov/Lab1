## Прапор Норвегії у форматі SVG

```python
originalNumber = input("Enter the number:")
originalNumber = int(originalNumber)
a = 1
b = 1
result = "This number isn't sum of squares."
while a < originalNumber:
    b = 1
    while b < a:
        if originalNumber == a * a + b * b:
            result = f'{a}^2+{b}^2={originalNumber}'
            break
        b += 1
    a += 1
print(result)
```
